```python
import numpy as np
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from scipy.stats import norm
yahoo_data = yf.download('^GSPC', start='1950-01-01', end='2023-02-20', interval='1d')

daily_return = yahoo_data.pct_change()

high_return = daily_return.loc[:,["Low"]]
t = np.linspace(high_return.min(), high_return.max(),500)
sb.histplot(high_return,stat="density")
plt.plot(t, norm.pdf(t, loc=high_return.mean(), scale=high_return.std()),
    label='Gaussian', c='orange')
# sb.lineplot(df,x='col1', y='col2')
# print(yahoo_data)
# print(daily_return.loc[:,["High"]])
```
